```python
av_daily_power = powgen.groupby('date').agg(np.mean)
```

```python
plt.figure(figsize=(15,5))
plt.plot(av_daily_power['thermal_generation_actual'],color='red' ,label ='Thermal')
plt.plot(av_daily_power['nuclear_generation_actual'],color='green', label ='nuclear')
plt.legend()
plt.title('Average daily generated')
plt.xlabel('date')
plt.ylabel('Power generated')
plt.show()
```
