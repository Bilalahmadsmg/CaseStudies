```python
for s in sample_string.split():
    print(s.strip(puncts))
```

```python
def remove_punctualtions(text):
    res = []
    for t in text.split():
        if t.lower().startswith("http"):
            res.append(t)
        else:
            res.append(t.translate(str.maketrans("", "", puncts)))
    return " ".join(res)
```

```python
remove_punctualtions(sample_string)
```

```python
data['text'] = data['text'].map(remove_punctualtions)
data.head()
```

### StopWords Removal

```python
from nltk.corpus import stopwords
```

```python
stw = stopwords.words("English")
stw
```

```python
def remove_stopwords(text):
    print (text)
    
    res = []
    
    for w in text.split():
        if w.lower not in stw:
            res.append(w)
            
    return " ".join(res)
```

```python
remove_stopwords('I am not coming home')
```

```python
data['text']
```

```python
data['text'] = data['text'].map(remove_stopwords)
```

```python
data.head()
```

### Removing Frequent words

```python
from collections import Counter
cnt = Counter()
```

```python
for text in data['text'].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)
```

```python
most_frequent = [c[0] for c in cnt.most_common(10)]
most_frequent
```

```python
def remove_frequent_words(text):
    res = []
    
    for w in text.split():
        if w.lower not in most_frequent:
            res.append(w)
    return " ".join(res)
```

```python
remove_frequent_words("Hello to the India for you as a For country")
```

```python
data('text') = data('text').map(remove_frequent_words)
data.head()
```

### Removal of most rare words

```python
cnt.most_common()[-11:-1]
```

```python
[w for w in cnt.most_common() if w[1] == 1]
```

```python
rare_words = set([w for w in cnt.most_common() if w[1] == 1])
rare_words
```

```python
def remove_rare_words(text):
    
    res = []
    
    for w in text.split():
        if w.lower() not in rare_words:
            res.append(w)
            
    return " ".join(res)
```

```python
remove_rare_words("I displaced 659693 for effect")
```

```python
data('text') = data('text').map(remove_rare_words)
```
